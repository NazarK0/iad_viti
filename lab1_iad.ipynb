# Виконання

1. Підключення бази даних

```python
import pandas as pd
import sqlite3

con = sqlite3.connect("material_service.sqlite")
```

### SQL запит

```python
invoices = pd.read_sql_query("SELECT \
    Inv.id as inv_id, Inv.*, \
    M.name, M.fk_company, \
    G.*, \
    U.title AS 'units', \
    T.*, \
    C.title AS 'company_title', C.fk_battalion, \
    B.title AS 'battalion_title' \
    FROM Invoices AS Inv \
    LEFT JOIN Goods AS G ON Inv.fk_goods=G.id \
    LEFT JOIN Times AS T ON Inv.fk_time=T.id \
    LEFT JOIN Military AS M ON Inv.fk_enlistee=M.id \
    LEFT JOIN Companys AS C ON M.fk_company=C.id \
    LEFT JOIN Battalions AS B ON C.fk_battalion=B.id \
    LEFT JOIN UnitsOfMeasurement AS U ON G.fk_unit_of_measurement=U.id \
    ", con)[
    ["inv_id", "month", "quarter", "year", "name", "company_title", "battalion_title", "title", "amount", "units"]
]
```

### Вивід даних(За вимогою)

```python
from IPython.core.display import display, HTML

display(HTML(invoices.to_html()))
```

### Формування зведеної таблиці

```python
import numpy as np

pd.pivot_table(invoices, 
              index=["title","battalion_title", "company_title", "name"],
              columns=["year", "quarter", "month"],
              values="amount",
              aggfunc=np.sum,
              fill_value=0,
              margins=True
              )
```
